# 维修工单系统 - Android APK 编译

本笔记本用于在 Google Colab 云端编译 Android APK。

## 使用方法
1. 上传项目文件到 `/content/workorder-app/` 目录
2. 依次运行以下单元格
3. 下载生成的 APK 文件

```
# 1. 安装系统依赖
!apt-get update -qq
!apt-get install -y -qq build-essential libsqlite3-dev zlib1g-dev \
    libncurses5-dev libncursesw5-dev libreadline-dev libssl-dev \
    libgdbm-dev libgdbm-compat-dev liblzma-dev libffi-dev \
    libbz2-dev uuid-dev libfreetype6-dev libsdl2-dev \
    libsdl2-image-dev libsdl2-mixer-dev libsdl2-ttf-dev \
    libportmidi-dev libswscale-dev libavformat-dev libavcodec-dev

print("✅ 系统依赖安装完成")
```

```
# 2. 安装 Python 依赖
!pip install -q buildozer cython==0.29.33 kivy requests pillow
print("✅ Python 依赖安装完成")
```

```
# 3. 上传项目文件
import os
import shutil
from google.colab import files

# 创建项目目录
os.makedirs('/content/workorder-app', exist_ok=True)
%cd /content/workorder-app

print("请上传以下文件:")
print("- main.py")
print("- config.py")
print("- api_client.py")
print("- buildozer.spec")
print("\n点击下方 'Choose Files' 上传...")

uploaded = files.upload()
print(f"\n✅ 已上传 {len(uploaded)} 个文件")
```

```
# 4. 编译 APK (首次运行约需 15-30 分钟)
%cd /content/workorder-app
!buildozer android debug

print("\n✅ 编译完成!")
```

```
# 5. 下载 APK
import glob
from google.colab import files as colab_files

%cd /content/workorder-app
apk_files = glob.glob('bin/*.apk')

if apk_files:
    print(f"找到 {len(apk_files)} 个 APK 文件:")
    for apk in apk_files:
        print(f"  - {apk}")
        colab_files.download(apk)
else:
    print("❌ 未找到 APK 文件，请检查编译日志")
```
